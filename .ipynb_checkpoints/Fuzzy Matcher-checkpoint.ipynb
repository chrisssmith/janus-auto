```python
print("Started OK")

def organise_emails_numerically():
    ''' Takes all emails in the data folder and renames them numerically incrementing. '''
    nameCount = 0
    root = 'data'
    for filename in os.listdir(root):
        if filename.endswith(".msg"):
            newName = str(nameCount) + ".msg"
            print(newName)
            try:
                os.rename(os.path.join(root, filename), os.path.join(root, newName))
            except WindowsError:
                pass
            nameCount = nameCount + 1
    print("Renamed all files")


def extract_data_from_emails():
    ''' Extracts all .msg files in the data folder and then moves the output folders into the data/Processed folder '''
    for message in glob.glob('./data/*.msg'):
        subprocess.call(["python", "ExtractMsg.py", "--json", os.path.basename(message)], cwd="data", shell=True)         #Calls the ExtractMsg script within the data folder to execute on each .msg file

    root, dirs, files = os.walk("data").next()                                      # Generates a file directory
    nameCount = 0

    for dir in dirs:                                                                # Iterates through all directories in data folder
        os.rename(os.path.join(root, dir), os.path.join(root, str(nameCount)))      # Renames the folder with a numerical name
        shutil.move(os.path.join(root, str(nameCount)), "data/Processed/")          # Moves the folder to the data/Processed folder
        nameCount = nameCount + 1

def parse_json_email(id):
    ''' Takes the ID of an email and parses the data of the email.'''
    regex = re.compile('[^a-zA-Z]')
    path = "data/Processed/" + str(id) + "/message.json"
    print("Parsing " + path)
    files_array = seek_through_emails(id)
    attachment_content = parse(id, files_array)
    try:
        with open(path) as json_data:
            try:
                parsed_json = json.load(json_data)
                #print parsed_json
                #print parsed_json['date']
                attachmentCount = 0
                if attachmentCount < len(parsed_json['attachments']):

                    attachmentCount = + 1


                attachmentCount = len(parsed_json['attachments'])
                print attachmentCount
                attachment_filetype = {}

                if attachmentCount > 0:
                    counter = 0
                    print "Attachments: " + str(attachmentCount)
                    while counter < attachmentCount:
                        print counter
                        print len(attachment_content)
                        attachment_filetype[counter] = identify_filetype(parsed_json['attachments'][counter])
                        try:
                            attachment_content[counter] = regex.sub(' ', attachment_content[counter])
                            attachment_content[counter] = attachment_content[counter].decode('utf-8', 'ignore').encode("utf-8")

                        except:
                            print "Caught it!"
                        counter = counter + 1


                    parsed_json['to'] = re.sub(r'[\x00-\x08\x0b\x0c\x0e-\x1f\x7f-\xff]', '', repr(parsed_json['to']))
                    try:

                        ws.append(
                            [parsed_json['date'], parsed_json['subject'], parsed_json['from'], parsed_json['to'],
                             parsed_json['cc'],
                             parsed_json['body'], len(parsed_json['attachments'])])

                        standard = [parsed_json['date'], parsed_json['subject'], parsed_json['from'], parsed_json['to'],
                             parsed_json['cc'],
                             parsed_json['body'], len(parsed_json['attachments'])]

                        count = 0
                        extra = []
                        while count < len(parsed_json['attachments']):
                            extra.append(parsed_json['attachments'][count])
                            extra.append(attachment_filetype[count])
                            extra.append(attachment_content[count])
                            count = count + 1
                        merged = standard + extra
                        ws.append(merged)
                    except IndexError:
                        print("ERROR3")
                    except openpyxl.utils.exceptions.IllegalCharacterError:
                        print("Illegal Character Error")

                        print(repr(parsed_json['to']))



            except ValueError:
                print("Something wrong json file!")
    except IOError:
        print("ERROR")



def write_csv():
    ''' This is just an example of how reading to a CSV file works '''
    with open('parsed_emails.csv', 'w') as csvfile:
        fieldnames = ['Date', 'Subject', 'Subject', 'Sender','Sent To', 'Attachments', 'Attachment1', 'Attachment1 Type', 'Attachment1 Contents'
            , 'Attachment2', 'Attachment2 Type', 'Attachment2 Contents', 'Attachment3', 'Attachment3 Type', 'Attachment3 Contents',
                      'Attachment4', 'Attachment4 Type', 'Attachment4 Contents', 'Attachment5', 'Attachment5 Type', 'Attachment5 Contents']
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)

        writer.writeheader()
        writer.writerow({'Test_Col1': '1', 'Test_Col2': '2'})
        writer.writerow({'Test_Col1': '3', 'Test_Col2': '4'})
        writer.writerow({'Test_Col1': '5', 'Test_Col2': '6'})


def seek_through_emails(id):
    '''This iterates through all processed email folders and the files within the folders.'''
    directory = "data/Processed/" + str(id) + "/"
    print(directory)
    try:
        folders =  os.walk(directory).next()[2]
        files_array = []
        for files in folders:
            files_array.append(files)
        return files_array
    except StopIteration:
        print("FAIL")
        print("FAILr")




def ocr(full_path):
    ''' Uses pytesseract to convert an image into text'''
    fullpath = full_path
    pytesseract.pytesseract.tesseract_cmd = 'C:/Program Files (x86)/Tesseract-OCR/tesseract'
    my_path = os.path.abspath(__file__)
    mydir = os.path.dirname(my_path)
    start = os.path.join(mydir, "img002.jpg")

    return pytesseract.image_to_string(Image.open(fullpath))


def parse_doc(full_path):
    ''' Uses antiword to convert a .doc file into text '''
    fullpath = "../" + full_path
    parsed_doc = subprocess.check_output(["antiword", "-t", fullpath], cwd="antiword", shell=True)
    return parsed_doc


def parse_docx(full_path):
    ''' Uses docx to convert a .docx file into text'''
    fullpath = "../" + full_path
    document = opendocx(full_path)
    paratextlist = getdocumenttext(document)
    newparatextlist = []
    for paratext in paratextlist:
        newparatextlist.append(paratext.encode("utf-8"))
    return '\n\n'.join(newparatextlist)


def parse_pdf(full_path):
    ''' Uses Pdfminer to convert a pdf to into text'''
    fullpath = "../" + full_path
    parsed_pdf = subprocess.check_output(["python", "tools/pdf2txt.py", fullpath], cwd="pdfminer", shell=True)
    return parsed_pdf

# Identifies the file type
# @param file_name <desc>
# @return <desc>
def identify_filetype(file_name):
    filename, file_extension = os.path.splitext(file_name)
    return file_extension

def parse(id,attachments):
    path = "data/Processed/" + str(id) + "/"
    data = []
    try:
        for attachment in attachments:
            full_path = path + str(attachment)
            if identify_filetype(attachment) == ".doc":
                data.append(parse_doc(full_path))
            if identify_filetype(attachment) == ".docx":
                data.append(parse_docx(full_path))
            if identify_filetype(attachment) == ".png" or identify_filetype(attachment) == ".jpg" \
            or identify_filetype(attachment) == ".jpeg":
                data.append(ocr(full_path))
            if identify_filetype(attachment) == ".pdf":
                data.append(parse_pdf(full_path))
    except TypeError:
        print("Error")
    print data
    return data


# This function opens up the worksheet.
def initalize_worksheet():
    global wb
    global ws
    wb = Workbook()
    ws = wb.active

    ws.append(['Date', 'Subject', 'Sender', 'Sent To', 'CC', 'Body' ,'Attachments', 'Attachment1', 'Attachment1 Type'
                 ,'Attachment1 Contents' , 'Attachment2', 'Attachment2 Type','Attachment2 Contents', 'Attachment3',
               'Attachment3 Type','Attachment3 Contents', 'Attachment4', 'Attachment4 Type','Attachment4 Contents',
               'Attachment5', 'Attachment5 Type','Attachment5 Contents', 'Attachment6', 'Attachment6 Type',
               'Attachment6 Contents', 'Attachment7', 'Attachment7 Type','Attachment7 Contents', 'Attachment8',
               'Attachment8 Type','Attachment8 Contents' ])

def close_save_worksheet():
    # Save the file
    wb.save("sample.xlsx")

def purge_processed():
    import os, shutil
    folder = 'data/Processed/'
    for the_file in os.listdir(folder):
        file_path = os.path.join(folder, the_file)
        try:
            if os.path.isfile(file_path):
                os.unlink(file_path)
            elif os.path.isdir(file_path): shutil.rmtree(file_path)
        except Exception as e:
            print(e)
def process_emails():
    list = os.listdir("data/Processed/") # dir is your directory path
    emails = len(list)
    count = 0
    while count < emails:
        print("Parsing Email "+str(count))
        parse_json_email(count)
        count = count +1
    print("Emails processed")

purge_processed()
initalize_worksheet()
#organise_emails_numerically()
extract_data_from_emails()
process_emails()
close_save_worksheet()
purge_processed()
#parse_json_email(6)

#write_csv()
#seek_through_emails()
#ocr()
#parse_doc()
#parse_docx()
#parse_pdf()
#print identify_filetype("test.jpg")
#print parse("test2")
```
